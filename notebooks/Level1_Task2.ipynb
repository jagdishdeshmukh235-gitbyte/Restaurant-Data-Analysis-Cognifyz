```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
plt.style.use("ggplot")
```

```python
df = pd.read_csv("../data/Dataset .csv")
```

```python
print("Dataset Shape :", df.shape)
```

```python
df.head()
```

```python
df.describe().T
```

```python
df.mean(numeric_only=True)
```

```python
df.median(numeric_only=True)
```

```python
df.std(numeric_only=True)
```

```python
country = df["Country Code"].value_counts()

country
```

```python
plt.figure(figsize=(10,5))

country.plot(kind="bar")

plt.title("Number of Restaurants by Country Code")
plt.xlabel("Country Code")
plt.ylabel("Restaurant Count")

plt.tight_layout()

plt.savefig("../images/country_distribution.png", dpi=300)

plt.show()
```

```python
city = df["City"].value_counts()

city.head(10)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x=city.head(10).index,
    y=city.head(10).values
)

plt.xticks(rotation=45)

plt.title("Top 10 Cities by Number of Restaurants")

plt.xlabel("City")

plt.ylabel("Restaurant Count")

plt.tight_layout()

plt.savefig("../images/top10_cities.png", dpi=300)

plt.show()
```

```python
cuisine = df["Cuisines"].value_counts()

cuisine.head(10)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x=cuisine.head(10).index,
    y=cuisine.head(10).values
)

plt.xticks(rotation=90)

plt.title("Top 10 Most Popular Cuisines")

plt.xlabel("Cuisine")

plt.ylabel("Restaurant Count")

plt.tight_layout()

plt.savefig("../images/top10_cuisines.png", dpi=300)

plt.show()
```
