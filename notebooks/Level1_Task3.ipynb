```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
plt.style.use("ggplot")
```

```python
df = pd.read_csv("../data/Dataset .csv")
```

```python
df[["Restaurant Name","City","Latitude","Longitude"]].head()
```

```python
plt.figure(figsize=(12,8))

plt.scatter(
    df["Longitude"],
    df["Latitude"],
    alpha=0.5,
    color="blue"
)

plt.title("Restaurant Locations")

plt.xlabel("Longitude")

plt.ylabel("Latitude")

plt.grid(True)

plt.savefig("../images/restaurant_locations.png", dpi=300)

plt.show()
```

```python
top_city = df["City"].value_counts().head(10)

top_city
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x=top_city.index,
    y=top_city.values
)

plt.xticks(rotation=45)

plt.title("Top 10 Cities by Number of Restaurants")

plt.xlabel("City")

plt.ylabel("Restaurant Count")

plt.tight_layout()

plt.savefig("../images/top10_city_distribution.png", dpi=300)

plt.show()
```

```python
rating_city = df.groupby("City")["Aggregate rating"].mean()

rating_city.sort_values(ascending=False).head(10)
```

```python
top_rating = rating_city.sort_values(
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_rating.index,
    y=top_rating.values
)

plt.xticks(rotation=45)

plt.title("Top 10 Cities with Highest Average Rating")

plt.xlabel("City")

plt.ylabel("Average Rating")

plt.tight_layout()

plt.savefig("../images/highest_rating_city.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.kdeplot(
    x=df["Longitude"],
    y=df["Latitude"],
    fill=True,
    cmap="Blues"
)

plt.title("Restaurant Density")

plt.xlabel("Longitude")

plt.ylabel("Latitude")

plt.tight_layout()

plt.savefig("../images/restaurant_density.png", dpi=300)

plt.show()
```
