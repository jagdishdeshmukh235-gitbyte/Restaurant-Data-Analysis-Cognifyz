```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)

plt.style.use("ggplot")

print("Libraries Imported Successfully!")
```

```python
df = pd.read_csv("../data/Dataset .csv")

print("Dataset Loaded Successfully!")
```

```python
df[[
    "Price range",
    "Aggregate rating",
    "Rating color"
]].head()
```

```python
price_range = df["Price range"].value_counts()

price_range
```

```python
most_common = df["Price range"].mode()[0]

print("Most Common Price Range :", most_common)
```

```python
plt.figure(figsize=(8,6))

sns.countplot(
    data=df,
    x="Price range",
    palette="viridis"
)

plt.title("Restaurant Distribution by Price Range")

plt.xlabel("Price Range")

plt.ylabel("Number of Restaurants")

plt.tight_layout()

plt.savefig("../images/price_range_distribution.png", dpi=300)

plt.show()
```

```python
avg_rating = df.groupby(
    "Price range"
)["Aggregate rating"].mean().round(2)

avg_rating
```

```python
plt.figure(figsize=(8,6))

sns.barplot(
    x=avg_rating.index,
    y=avg_rating.values,
    palette="magma"
)

plt.title("Average Rating by Price Range")

plt.xlabel("Price Range")

plt.ylabel("Average Rating")

plt.tight_layout()

plt.savefig("../images/average_rating_price_range.png", dpi=300)

plt.show()
```

```python
df["Rating color"].value_counts()
```

```python
rating_color = df.groupby(
    "Rating color"
)["Aggregate rating"].mean().round(2)

rating_color.sort_values(
    ascending=False
)
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x=rating_color.index,
    y=rating_color.values,
    palette="Set2"
)

plt.title("Average Rating by Rating Color")

plt.xlabel("Rating Color")

plt.ylabel("Average Rating")

plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig("../images/rating_color_analysis.png", dpi=300)

plt.show()
```

```python
highest_color = rating_color.idxmax()

highest_rating = rating_color.max()

print("Highest Rating Color :", highest_color)

print("Average Rating :", highest_rating)
```
