```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)

plt.style.use("ggplot")

print("Libraries Imported Successfully!")
```

```python
df = pd.read_csv("../data/Dataset_Feature_Engineered.csv")

df.head()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x="Aggregate rating",
    bins=20,
    color="royalblue"
)

plt.title("Distribution of Aggregate Ratings")

plt.xlabel("Aggregate Rating")

plt.ylabel("Restaurant Count")

plt.tight_layout()

plt.savefig("../images/rating_distribution_dashboard.png",dpi=300)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(
    data=df,
    x="Aggregate rating",
    palette="viridis"
)

plt.xticks(rotation=45)

plt.title("Aggregate Rating Count")

plt.tight_layout()

plt.savefig("../images/rating_count.png",dpi=300)

plt.show()
```

```python
top_city = df["City"].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_city.index,
    y=top_city.values,
    palette="magma"
)

plt.xticks(rotation=45)

plt.title("Top 10 Cities")

plt.xlabel("City")

plt.ylabel("Restaurant Count")

plt.tight_layout()

plt.savefig("../images/dashboard_top_city.png",dpi=300)

plt.show()
```

```python
top_cuisine = df["Cuisines"].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_cuisine.values,
    y=top_cuisine.index,
    palette="coolwarm"
)

plt.title("Top 10 Cuisines")

plt.xlabel("Restaurant Count")

plt.ylabel("Cuisine")

plt.tight_layout()

plt.savefig("../images/dashboard_top_cuisine.png",dpi=300)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="Votes",
    y="Aggregate rating",
    alpha=0.6,
    color="red"
)

plt.title("Votes vs Aggregate Rating")

plt.tight_layout()

plt.savefig("../images/votes_vs_rating_dashboard.png",dpi=300)

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    data=df,
    x="Price range",
    y="Aggregate rating",
    palette="Set2"
)

plt.title("Price Range vs Aggregate Rating")

plt.tight_layout()

plt.savefig("../images/price_vs_rating.png",dpi=300)

plt.show()
```

```python
numeric_df = df.select_dtypes(include=["number"])
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.tight_layout()

plt.savefig("../images/correlation_heatmap.png",dpi=300)

plt.show()
```

```python
sns.pairplot(

    df[[
        "Aggregate rating",
        "Votes",
        "Average Cost for two",
        "Price range"
    ]]

)

plt.savefig("../images/pairplot.png",dpi=300)

plt.show()
```

```python
plt.figure(figsize=(6,6))

df["Has Online delivery"].value_counts().plot(

    kind="pie",

    autopct="%1.1f%%"

)

plt.ylabel("")

plt.title("Online Delivery")

plt.savefig("../images/online_delivery_dashboard.png",dpi=300)

plt.show()
```

```python
plt.figure(figsize=(6,6))

df["Has Table booking"].value_counts().plot(

    kind="pie",

    autopct="%1.1f%%"

)

plt.ylabel("")

plt.title("Table Booking")

plt.savefig("../images/table_booking_dashboard.png",dpi=300)

plt.show()
```
