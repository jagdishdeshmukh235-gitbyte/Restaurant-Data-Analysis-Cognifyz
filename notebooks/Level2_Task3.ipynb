```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)

plt.style.use("ggplot")

print("Libraries Imported Successfully!")
```

```python
df = pd.read_csv("../data/Dataset .csv")

print("Dataset Loaded Successfully!")
```

```python
df.head()
```

```python
df[[
    "Restaurant Name",
    "Address",
    "Has Table booking",
    "Has Online delivery"
]].head()
```

```python
df["Restaurant_Name_Length"] = df["Restaurant Name"].str.len()

df[["Restaurant Name","Restaurant_Name_Length"]].head()
```

```python
df["Address_Length"] = df["Address"].str.len()

df[["Address","Address_Length"]].head()
```

```python
df["Table_Booking"] = df["Has Table booking"].map({
    "Yes":1,
    "No":0
})

df[["Has Table booking","Table_Booking"]].head()
```

```python
df["Online_Delivery"] = df["Has Online delivery"].map({
    "Yes":1,
    "No":0
})

df[["Has Online delivery","Online_Delivery"]].head()
```

```python
df[[
    "Restaurant_Name_Length",
    "Address_Length",
    "Table_Booking",
    "Online_Delivery"
]].head()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df["Restaurant_Name_Length"],
    bins=20,
    color="royalblue"
)

plt.title("Distribution of Restaurant Name Length")

plt.xlabel("Restaurant Name Length")

plt.ylabel("Count")

plt.tight_layout()

plt.savefig("../images/restaurant_name_length.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df["Address_Length"],
    bins=20,
    color="green"
)

plt.title("Distribution of Address Length")

plt.xlabel("Address Length")

plt.ylabel("Count")

plt.tight_layout()

plt.savefig("../images/address_length_distribution.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(
    x="Table_Booking",
    data=df
)

plt.title("Encoded Table Booking")

plt.tight_layout()

plt.savefig("../images/table_booking_encoded.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(
    x="Online_Delivery",
    data=df
)

plt.title("Encoded Online Delivery")

plt.tight_layout()

plt.savefig("../images/online_delivery_encoded.png", dpi=300)

plt.show()
```

```python
df.to_csv("../data/Dataset_Feature_Engineered .csv", index=False)

print("New Dataset Saved Successfully!")
```
